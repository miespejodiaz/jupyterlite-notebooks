```python
# Calculate enrichment of 6mA in chromatin stenciling experiments

This small project contains scripts and resources for calculating the peak over background levels of 6mA in chromatin stenciling (fiber-seq) experiments.
To perform the evaluation, there are three steps:

## Quick start

0. Install a python environment based on the requirements in `requirements.txt`, there is a helper script (`install_venv.sh`) that you may use.
1. Run `$ bash run_6mA_peak_over_background.sh` running this script with no arguments will display a help and description of the input files required.
2. Run `$ python calculate_snr.py ${localize_tsv} [--outdir ${output_directory}]` to calculate the metric. Optionally output a new plot.

## Description of metric

The output metric peak_over_background the ratio of the 6mA levels centered at housekeeping gene promoters. 
The housekeeping genes are culled from the [HRT atlas](https://housekeeping.unicamp.br/) then joined with promoters from the [Eukaryotic promoter database](https://epd.expasy.org/epd).

Below is an example plot generated by `modkit localize`:

![signal-to-noise](signal-to-noise-desc.png)

The metric is then a ratio of the value of the blue line and the red line.

## Notes

- For Human runs, align your reads to hg38
- For Mouse runs, align your reads to mm10
- Use `dorado aligner` to map reads (that aren't mapped during basecalling) to make sure that modified base tags aren't corrupted or lost
```
